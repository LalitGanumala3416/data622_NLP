```python
import benepar
benepar.download("benepar_en3")
```

```python
import spacy
import nltk
from nltk import Tree

nlp = spacy.load("en_core_web_sm")

# ADD THIS LINE HERE
nlp.add_pipe("benepar", config={"model": "benepar_en3"})
```

```python
# Download punkt if needed
nltk.download("punkt")

# Gettysburg Address – first two sentences
s1 = ("Four score and seven years ago our fathers brought forth on this continent, "
      "a new nation, conceived in Liberty, and dedicated to the proposition that all men are created equal.")

s2 = ("Now we are engaged in a great civil war, testing whether that nation, or any nation so conceived "
      "and so dedicated, can long endure.")

doc1 = nlp(s1)
doc2 = nlp(s2)
```

```python
# Tokenization
print("1) spaCy Tokens - Sentence 1:")
print([token.text for token in doc1])

print("\n1) spaCy Tokens - Sentence 2:")
print([token.text for token in doc2])

# benepar parse strings
sent1 = list(doc1.sents)[0]
sent2 = list(doc2.sents)[0]

print("\n1) benepar Parse - Sentence 1:")
print(sent1._.parse_string)

print("\n1) benepar Parse - Sentence 2:")
print(sent2._.parse_string)
```

```python
# Part-of-Speech Tagging
print("\n2) POS Tags - Sentence 1:")
for token in doc1:
    print(f"{token.text:12} POS={token.pos_:6} TAG={token.tag_}")
```

```python
# Dependency Parsing
print("\n3) Dependency Parsing - Sentence 2:")
for token in doc2:
    print(f"{token.text:12} dep={token.dep_:10} head={token.head.text}")
```

```python
# Constituent Parsing
print("\n4) Constituency Parse Tree - Sentence 1:")
tree = Tree.fromstring(sent1._.parse_string)
tree.pretty_print()
```

```python
# Extract Noun Phrases
print("\n5) Noun Chunks - Sentence 1:")
for chunk in doc1.noun_chunks:
    print("-", chunk.text)

print("\n5) Noun Chunks - Sentence 2:")
for chunk in doc2.noun_chunks:
    print("-", chunk.text)
```
