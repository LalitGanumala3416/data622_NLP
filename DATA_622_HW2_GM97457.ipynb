```python
import requests
from bs4 import BeautifulSoup
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# 0. One-time NLTK downloads (run once)
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("wordnet")

url = "https://www.cnn.com/2025/06/13/style/why-luxury-brands-are-so-expensive"
headers = {
    "User-Agent": "DATA622-homework-bot/1.0 (gm97457@umbc.edu)"
}
```

```python
# 1. Read the file and print first 700 characters
resp = requests.get(url, headers=headers)
html = resp.text
print("1) First 700 characters of raw HTML:\n")
print(html[:700])
```

```python
# 2. Remove HTML tags – keep only raw text
soup = BeautifulSoup(html, "html.parser")
raw_text = soup.get_text(separator=" ")
raw_text = re.sub(r"\s+", " ", raw_text).strip()
print("\n2) First 700 characters of plain text:\n")
print(raw_text[:700])
```

```python
# 3. Lowercase and remove punctuation
lower_text = raw_text.lower()
translator = str.maketrans("", "", string.punctuation)
no_punct_text = lower_text.translate(translator)
print("\n3) First 700 characters after lowercase + no punctuation:\n")
print(no_punct_text[:700])
```

```python
# 4. Remove English stopwords
tokens = nltk.word_tokenize(no_punct_text)
stop_words = set(stopwords.words("english"))
tokens_no_stop = [w for w in tokens if w not in stop_words and w.isalpha()]
print("\n4) First 50 tokens after stopword removal:\n")
print(tokens_no_stop[:50])
```

```python
# 5. Lemmatize the text
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(w) for w in tokens_no_stop]
print("\n5) First 50 lemmatized words:\n")
print(lemmas[:50])
```

```python
# 6. Comparing it to stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stems = [stemmer.stem(w) for w in tokens_no_stop]
print("\n5b) First 50 stemmed words (for comparison):\n")
print(stems[:50])
```

Yes, there is a difference. Lemmatization retains the words in their correct dictionary forms, and stemming usually cuts off endings to form shorter and even non-words roots. Through lemmatization, as compared to stemming, provides readable language such as "luxury, expensive, loaded, moved, repetitive" instead of the variants such as "luxuri, expens, load, move, repetit" (which are non-English words). This indicates that lemmatization is less abstract and more readable and stemming is a more raw and rule-based truncation.
