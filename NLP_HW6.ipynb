```python
import random
import re

# 1. Sample training text (you can replace this with a longer fairy tale)
training_text = """
Once upon a time there was a kingdom hidden between blue mountains and a wide, whispering sea.
In this kingdom lived a gentle queen who spent her days listening to her people’s hopes and worries.
One day, she discovered a dusty map tucked inside an old music box in the palace attic.
Beyond the edge of the map lay a lost valley said to hold a spring of endless courage and kindness.
The queen swore to find this valley and share its blessings with every village beneath the castle walls.
"""

# Clean and split into words
def tokenize(text):
    # keep simple word tokens, strip punctuation around them
    tokens = re.findall(r"\b\w+\b", text)
    return tokens

words = tokenize(training_text)

# 2. Build Markov chain: map each word to list of possible next words
markov_chain = {}
for i in range(len(words) - 1):
    w, nxt = words[i], words[i+1]
    markov_chain.setdefault(w, []).append(nxt)

# 3. Generate story starting with given phrase
start_phrase = "Once upon a time there was a kingdom"
start_tokens = tokenize(start_phrase)

# last word of the given phrase will be our starting key
current_word = start_tokens[-1]

story_tokens = start_tokens[:]  # begin with provided phrase
desired_length = 150  # total number of words in generated story (adjust as you like)

for _ in range(desired_length - len(story_tokens)):
    next_candidates = markov_chain.get(current_word)
    if not next_candidates:
        # if we hit a word with no outgoing edge, restart from a random word
        current_word = random.choice(words)
        continue
    next_word = random.choice(next_candidates)
    story_tokens.append(next_word)
    current_word = next_word

story = " ".join(story_tokens)
print("Exercise 1 – Markov chain story:\n")
print(story)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = (
    "Write a creative short story about why it is important to study data science today. "
    "The story must have: 1) a main character, 2) a problem, and 3) how data science helps "
    "solve that problem. Do not explain what you are doing, just tell the story.\n\nStory:"
)

output = generator(
    prompt,
    max_new_tokens=350,
    do_sample=True,
    temperature=0.9,
    top_p=0.9,
    num_return_sequences=1,
)

story = output[0]["generated_text"].replace(prompt, "").strip()
print("\nExercise 2 – LLM-generated story:\n")
print(story)
```
